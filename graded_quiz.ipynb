```python
import pandas as pd 
import numpy as np 
import seaborn  as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df=pd.read_csv('Downloads\FoodBalanceSheets_E_Africa_NOFLAG.csv',encoding="latin-1")
```

```python


protein_supply_2015 = df[(df['Area'] == 'Madagascar') & (df['Element'] == 'Protein supply quantity (g/capita/day)')]['Y2015']
total_protein_supply_2015 = protein_supply_2015.sum()

print("Total Protein supply quantity in Madagascar in 2015:", total_protein_supply_2015)
```

```python


processing_sum_2017 = df[df['Element'] == 'Processing'].groupby('Element')['Y2017'].sum()

print("Total sum of Processing in 2017:", processing_sum_2017)
```

```python


sum_2017_by_area = df.groupby('Area')['Y2017'].sum()
area_highest_sum_2017 = sum_2017_by_area.idxmax()

print("Area with the highest sum in 2017:", area_highest_sum_2017)
```

```python

total_unique_countries = df['Area'].nunique()

print("Total number of unique countries in the dataset:", total_unique_countries)
```

```python


missing_data_2014 = df['Y2014'].isnull().sum()
total_data_2014 = len(df['Y2014'])
percentage_missing_2014 = (missing_data_2014 / total_data_2014) * 100

print("Total number of missing data in 2014:", missing_data_2014)
print("Percentage of missing data in 2014:", "{:.3f}%".format(percentage_missing_2014))
```

```python


grouped_area_sum_2017 = df.groupby('Area')['Y2017'].sum()
sorted_area_sum_2017 = grouped_area_sum_2017.sort_values()
seventh_lowest_area = sorted_area_sum_2017.index[6]

print("Area with the 7th lowest sum in 2017:", seventh_lowest_area)
```
